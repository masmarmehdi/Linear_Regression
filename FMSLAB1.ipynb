```python
# Values of x and y concerning to my number of variant.
x = [6.0, 6.2, 6.4, 6.6, 6.8, 7.0, 7.2, 7.4]
y = [8.78, 8.45, 7.98, 8.05, 8.01, 8.00, 7.97, 7.99]

# Creating an empty array to input the values of linear equation to draw the regression line
newArr = []
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
for i in range(len(x)):
    y1 = -0.493* x[i] + 11.45
    newArr.append(y1)

plt.plot(x, newArr, color ="grey", label = 'Linear Regression line')
ax.scatter(x, newArr, color = 'green', marker = '+', label = 'Estimated points')
ax.scatter(x, y, color='b', marker ='*', label = 'Coordinates of my variant')
ax.set_title('scatter diagram variant 10')
plt.xlabel("X axes", color = 'b')
plt.ylabel("Y axes", color = 'r')
averageOfX = 6.7
averageOfY = 8.15

ax.scatter(averageOfX, averageOfY, color = 'r', marker = '*' , label = 'Mean coordinate')
plt.legend()
plt.show()
```
